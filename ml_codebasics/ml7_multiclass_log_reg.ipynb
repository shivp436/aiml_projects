# IRIS Flower Classification

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Use seaborn's scatterplot function for hue-based plotting
sns.scatterplot(x=iris.data[:, 0], y=iris.data[:, 1],
                hue=iris.target, palette='viridis',
                legend='full')  # Show the legend

# Get the current legend handles and labels
handles, labels = plt.gca().get_legend_handles_labels()

# Replace numerical labels with target_names
plt.legend(handles, iris.target_names, title='Species')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

# Use seaborn's scatterplot function for hue-based plotting
sns.scatterplot(x=iris.data[:, 2], y=iris.data[:, 3],
                hue=iris.target, palette='viridis',
                legend='full')  # Show the legend

# Get the current legend handles and labels
handles, labels = plt.gca().get_legend_handles_labels()

# Replace numerical labels with target_names
plt.legend(handles, iris.target_names, title='Species')

plt.show()
```

From both the plots it is evident that
- Setosa
  - small petal width, petal length
  - small sepal length, and larger sepal width compared to other 2
- versicolor is in between
- viriginca is the largest species

```python
from sklearn.linear_model import LogisticRegression

species = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
species.fit(X_train, y_train)
```

```python
species.predict(X_test)
```

```python
species.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, species.predict(X_test))
cm
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

We can see that our model made a mistake when predicting 1 case. Truth was Species 2, while it predicted to be Species 1.

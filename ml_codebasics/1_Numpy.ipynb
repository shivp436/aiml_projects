```python
mat1 = [1, 2, 3]
mat2 = [4, 5, 6]
```

```python
def lin_multiply(mat1, mat2):
  result = []
  for x,y in zip(mat1, mat2):
    result.append(x*y)
  return result
```

```python
res = lin_multiply(mat1, mat2)
```
